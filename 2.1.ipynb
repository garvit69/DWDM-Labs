```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read excel file into a DataFrame
df = pd.read_excel("/content/drive/MyDrive/titanic.xls")
```

```python
# a.) histogram using singleton bucket

# Plot a histogram of the 'Age' attribute
plt.hist(df['age'], bins = 10)
plt.xlabel('age')
plt.ylabel('Frequency')
plt.title('Histogram of age')
plt.show()
```

```python
# Plot a histogram of the 'Age' attribute
plt.hist(df['fare'], bins = 10)
plt.xlabel('fare')
plt.ylabel('Frequency')
plt.title('Histogram of fare')
plt.show()
```

```python
# b.) Equal-width histogram of width = 10

# Plot a histogram of the 'Age' attribute
plt.hist(df['age'], bins = range(0, int(df['age'].max())+10, 10))
plt.xlabel('age')
plt.ylabel('Frequency')
plt.title('Histogram of age')
plt.show()
```

```python
# Plot a histogram of the 'Fare' attribute
plt.hist(df['fare'], bins = range(0, int(df['fare'].max())+10, 10))
plt.xlabel('fare')
plt.ylabel('Frequency')
plt.title('Histogram of age')
plt.show()
```
