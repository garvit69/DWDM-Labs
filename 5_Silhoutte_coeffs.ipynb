```python
import numpy as np
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
```

```python
# 1st set of clusters
cluster1 = np.array([(0,0), (0,1), (2,3)])
cluster2 = np.array([(3,3), (3,4)])
```

```python
data1 = np.concatenate((cluster1, cluster2), axis = 0)
labels1 = [0] * len(cluster1) + [1] * len(cluster2)
silhouette_avg1 = silhouette_score(data1, labels1, metric = 'euclidean')
```

```python
# 2nd set of clusters
cluster1 = np.array([(0,0), (0,1), (2,3)])
cluster2 = np.array([(3,3)])
cluster3 = np.array([(3,4)])
```

```python
data2 = np.concatenate((cluster1, cluster2, cluster3), axis = 0)
labels2 = [0] * len(cluster1) + [1] * len(cluster2) + [2] * len(cluster3)
silhouette_avg2 = silhouette_score(data2, labels2, metric = 'euclidean')
```

```python
print("Silhouette coefficient for 2 cluster set ==> ", silhouette_avg1)
print("Silhouette coefficient for 3 cluster set ==> ", silhouette_avg2)
```
