```python
# Min - max normalization

import pandas as pd
# Read the csv file into a DataFrame 
df = pd.read_excel("/content/drive/MyDrive/titanic.xls")
# Perform min-max normalization for the 'Age' attribute 
age_min = df['age'].min()
age_max = df['age'].max()
df['age'] = (df['age'] - age_min) / (age_max - age_min)
# Perform min-max normalization for the 'Fare' attribute fare_min = df['fare'].min()
fare_max = df['fare'].max()
fare_min = df['fare'].min()
df['fare'] = (df['fare'] - fare_min) / (fare_max - fare_min) 
print(df[['age','fare']])
```

```python
# Z - score normalization

import pandas as pd
titanic_df = pd.read_excel('/content/drive/MyDrive/titanic.xls')
# Calculate the mean and standard deviation for the "age" attribute 
age_mean = titanic_df['age'].mean()
age_std = titanic_df['age'].std()
# Calculate the mean and standard deviation for the "fare" attribute 
fare_mean = titanic_df['fare'].mean()
fare_std = titanic_df['fare'].std()
titanic_df['age_zscore'] = (titanic_df['age'] - age_mean) / age_std 
titanic_df['fare_zscore'] = (titanic_df['fare'] - fare_mean) / fare_std
# Print the first few rows of the DataFrame to verify the results 
print(titanic_df.head())
```
