```python
import pandas as pd
```

```python
df = pd.read_csv('basket_analysis.csv')
df.head()
```

```python
df = df.drop(columns=df.columns[0], axis=1)
df = df.replace({True:1, False:0})
df.head()
```

```python
min_supp = int(input("Enter minimumu support: "))
freq = pd.DataFrame({'item_set': df.sum().index, 'frequency':df.sum().values})
freq
```

```python
freq['item_set'] = freq['item_set'].apply(lambda x:x.split(","))
maxlen = 0
if((min_supp*df.shape[0])/100 <= freq['frequency'].max()):
  freq = freq[freq['frequency'] >= (min_supp*df.shape[0])/100]
  freq.reset_index(drop=True, inplace=True)
  while(freq.shape[0]):
    maxlen += 1
    print(str(maxlen)+ " length item set")
    print(freq)
    newfreq = pd.DataFrame(columns=['item_set', 'frequency'])
    for i in freq.index:
      for j in range(i+1, freq.shape[0]):
        if(freq.iloc[i].item_set[:-1] == freq.iloc[j].item_set[:-1]):
          next_item_set = freq.iloc[i].item_set.copy()
          next_item_set.append(freq.iloc[j].item_set[-1])
          frequency = 0
          for k in df.index:
            flag=True
            for l in next_item_set:
              if(df.iloc[k][l] == 0):
                flag=False
                break
            if(flag):
              frequency +=1
          newfreq.loc[len(newfreq.index)] = [next_item_set, frequency]
    freq = newfreq[newfreq['frequency'] >= (min_supp*df.shape[0])/100]
  print("length of largest frequent of item set is " + str(maxlen))
```
