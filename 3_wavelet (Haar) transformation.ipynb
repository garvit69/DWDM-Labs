```python
import pywt
import numpy as np
import matplotlib.pyplot as plt
```

```python
ip_data = input("Enter input data in CSV format: ")
input_data = np.array([float(x.strip()) for x in ip_data.split(",")])
```

```python
coeffs = pywt.wavedec(input_data, 'haar', level = 1)
threshold = float(input("Enter the threshold value: "))
coeffs_threshold = pywt.threshold(coeffs, threshold, 'soft')
rec_signal = pywt.upcoef('a', coeffs_threshold[0], 'haar', level = 1)
```

```python
print("Original data ==> ", input_data)
print("Transformed coefficients ==> ", coeffs)
print("Reconstructed signal ==> ", rec_signal)
```

```python
plt.figure(figsize = (12,6))
plt.subplot(131)
plt.plot(input_data)
plt.title("Original Signal")
plt.subplot(132)
plt.plot(coeffs[0], label = "Approximation Coefficients")
plt.plot(coeffs[1], label = "Detail Coefficients")
plt.legend()
plt.title("Wavelet Transform")
plt.subplot(133)
plt.plot(rec_signal)
plt.title("Reconstructed Signal")
```
